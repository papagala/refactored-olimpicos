```python
import pandas as pd
```

```python
olympics_complete=pd.read_csv('../intermediate/olimpics_gdp_population.csv')
base_data_male = olympics_complete.loc[olympics_complete['Sex'] == "M",
                                              ['Sport',
                                               'Age',
                                               'Weight',
                                               'Height',
                                               'Medal_Won',
                                               'Event_Category',
                                               'GDP',
                                               'Population',
                                               'Team']].drop_duplicates()
```

```python
olympics_complete.columns
```

```python
no_medal_mask=base_data_male['Medal_Won']==0
medal_mask = [not i for i in no_medal_mask]
```

```python
medal_winner=base_data_male[medal_mask].copy()
medal_winner['Medal_Won_Corrected'] = medal_winner['Medal_Won']/medal_winner['Event_Category']
```

```python
medal_winner=base_data_male[medal_mask][['Age','Sport','Weight','Height','GDP','Population']]
```

```python
medal_winner_grouped=medal_winner.groupby(by=['Sport']).mean()
medal_winner_grouped['Won']=1
medal_winner_grouped
```

```python
no_medal_winner=base_data_male[no_medal_mask][['Age','Sport','Weight','Height','GDP','Population']]
```

```python
no_medal_winner_grouped=no_medal_winner.groupby(by='Sport').mean()
no_medal_winner_grouped['Won']=0
no_medal_winner_grouped
```

```python
final=medal_winner_grouped.append(no_medal_winner_grouped)
final['sport']=final.index
```
