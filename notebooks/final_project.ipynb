# Importar datos de olímpicos

```python
import pandas as pd

olympics_complete = pd.read_csv("../intermediate/olimpics_gdp_population.csv")
base_data_male = olympics_complete.loc[
    olympics_complete["Sex"] == "M",
    [
        "Sport",
        "Age",
        "Weight",
        "Height",
        "Medal_Won",
        "Event_Category",
        "GDP",
        "Population",
        "Team",
    ],
].drop_duplicates()
```

```python
olympics_complete.columns
```

```python
no_medal_mask = base_data_male["Medal_Won"] == 0
medal_mask = [not i for i in no_medal_mask]
```

```python
medal_winner = base_data_male[medal_mask].copy()
medal_winner["Medal_Won_Corrected"] = (
    medal_winner["Medal_Won"] / medal_winner["Event_Category"]
)
```

```python
base_data_male
```

```python
medal_winner = base_data_male[medal_mask][
    ["Age", "Sport", "Weight", "Height", "GDP", "Population", "Team"]
]
```

```python
medal_winner_grouped = medal_winner.groupby(by=["Sport"]).mean()
medal_winner_grouped["BMI"] = medal_winner_grouped["Weight"] / (
    (medal_winner_grouped["Height"] / 100) ** 2
)
medal_winner_grouped
```

## Algoritmo de clasificación

Finalmente tenemos el algoritmo de vecinos cercanos para buscar el vector de índice de masa corporal y altura de los medallistas olímpicos (adultos) históricos, que servirá como el IMC y altura ideales. Buscaremos dentro de todos los paises del mundo, aquellos niños entre 14-18 años que en promedio se acerquen más a este ideal, lo cual los coloca como paises potencia en la disciplina seleccionada (ya que incluso como adolescentes, cuentan en promedio con las cualidades físicas para obtener una medalla (inclusive en las olimpiadas de adultos, aunque el argumento se hace dentro del universo de los olímpicos juveniles)

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors


def buscar_vecinos_cercanos(deporte: str, nearest:int=3):
    """Función que busca los vecinos más cercanos al estandar ideal marcado por
    los medallistas olímpicos de toda la historia. Es decir, que paises poseen cualidades
    físicas más parecidas a los adultos que ganan medallas como adultos.

    Args:
        deporte (str): Deporte olímpico en inglés. Por ejemplo Handball, Hockey o Diving.
        nearest (int): Número de vectores más cercanos (3 por defecto).

    Returns:
        tuple: Regresa el una tupla donde el primer elemento
        es el resultado de NearestNeighbors.kneighbors. El segundo elemento es el promedio
        ideal de los medallistas olímpicos históricos para ayudar a explicar el resultado
    """
    matrix = np.matrix(imc_altura_boys.loc[:, ["Mean BMI", "Mean height"]])
    search_vec = np.matrix(
        medal_winner_grouped[["BMI", "Height"]].loc[
            deporte,
        ]
    )
    nbrs = NearestNeighbors(n_neighbors=nearest).fit(matrix)
    res = nbrs.kneighbors(search_vec)
    return res, search_vec
```

Mandamos llamar la función, y elegimos "Diving" para ver que obtenemos

```python
res = buscar_vecinos_cercanos("Diving",5)
# El primer argumento de res es la salida de NearestNeighbors.kneighbors
# que a su vez regresa una tupla donde lo que nos interesa es el segundo elemento (los índices)
# que podemos cruzar con el dataframe original imc_altura_boys
imc_altura_boys.loc[
    res[0][1][0],
]
```

El segundo elemento de la tupla nos sirve para explicar los resultados, ya que representa el ideal de los medallistas históricos
